```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
dir(iris)
```

```python
iris.feature_names, iris.target_names
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df
```

```python
df['target'] = iris.target
df
```

```python
df['target_name'] = df.target.apply( lambda x: iris.target_names[x] )
df.head()
df[df['target_name'] == 'setosa'].head()
```

```python
df0 = df[:50]
df1 = df[50:100]
df2 = df[100:150]
```

```python
from sklearn.datasets import load_digits
from sklearn.metrics import classification_report

digit = load_digits()
dir(digit)
```

```python
df = pd.DataFrame(digit.data, columns = digit.feature_names )
df.head()
```

```python
df["target"] = digit.target
df
```

```python
df[df.target==0].shape
```

```python
# Load the dataset
digits = load_digits()
X = digits.data  # Features (8x8 images flattened to 64 pixels)
y = digits.target  # Labels (0-9)

# Filter only digits 0 and 1
mask = (y == 0) | (y == 1)
X_filtered = X[mask]
y_filtered = y[mask]
```

```python
# Plot 5 samples of each digit
fig, axes = plt.subplots(2, 5, figsize=(10, 4))
for i in range(5):
    # Digit 0
    axes[0, i].imshow(X_filtered[y_filtered == 0][i].reshape(8, 8), cmap='gray')
    axes[0, i].set_title("Digit 0")
    axes[0, i].axis('off')
    
    # Digit 1
    axes[1, i].imshow(X_filtered[y_filtered == 1][i].reshape(8, 8), cmap='gray')
    axes[1, i].set_title("Digit 1")
    axes[1, i].axis('off')
plt.show()
```

```python
X = df.drop('target',axis=1)
y = df.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
len(X), len(X_train), len(X_test)
```

```python
model = SVC(kernel="rbf", C=1, )
model.fit(X_train, y_train)
model.score(X_test, y_test)
```
